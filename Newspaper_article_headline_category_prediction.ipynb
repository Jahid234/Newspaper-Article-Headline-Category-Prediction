```python
df = pd.read_json('News_Category_Dataset_v2.json', lines=True)
df
```

```python
df = df.drop(columns=['authors','link','short_description','date'])
df
```

```python
# Replace 'arts & culture' with 'culture and arts'

df['category'] = df['category'].replace('ARTS & CULTURE', 'CULTURE & ARTS')
df['category'] = df['category'].replace('PARENTING', 'PARENTS')
df['category'] = df['category'].replace('THE WORLDPOST', 'WORLDPOST')

print(df['category'])

len(df.category.unique())
```

```python
#convert all headlines to lowercase

df['headline'] = df['headline'].str.lower()
df
df.head(10)
```

```python
#splitting data into training and test set
X = df['headline']
Y = df['category']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=100, test_size=0.07, shuffle=True)
```

```python
y_test
```

```python
class_count = dict(y_train.value_counts())
class_count
```

```python
total_classes = len(y_train)
total_classes
```

```python
from nltk.corpus import stopwords
stop_words= set(stopwords.words('english'))
```

```python
#list of common words to be removed

import re
#Remove common words from the 'train_set'

for word in stop_words:
    X_train = X_train.str.replace(r'\b' + re.escape(word) + r'\b', '', regex=True).str.strip()

X_train
```

```python
#remove common words from test set

for word in stop_words:
    X_test = X_test.str.replace(r'\b' + re.escape(word) + r'\b', '', regex=True).str.strip()

X_test
```

```python
#finding probability of each class out of total classes

class_probability = {}

for key,value in class_count.items():
    class_probability[key] = value/total_classes

class_probability
```

```python
sum(class_probability.values())
```

```python
#storing all the words in each class separately in a dictionary

filtered_classes = {}
new = pd.concat([y_train, X_train], axis=1)

for key,value in class_count.items():
    s = ''
    x = new.loc[df['category'] == key, 'headline']
    for item in x:
        s += item
    s = s.split()
    filtered_classes[key] = s

filtered_classes['SPORTS']
```

```python
#finding no. of total unique words in the
total_words = []
for key, value in filtered_classes.items():
    total_words += value
```

```python
len(total_words)
```

```python
total_unique_words = set(total_words)
len(total_unique_words)
```

```python
# Applying Naive Bayes formula to the training dataset
def findAnswer(item):
    probabilites = {}
    test_string = item
    test_list = test_string.split()

    probability_list = []

    for i in test_list:
        for key, value in filtered_classes.items():
            count = 0
            for item in value:
                if i==item:
                    count += 1
            total_class_words = len(value)
            probabilites[key] = (count+1)/(total_class_words + len(total_unique_words))
        probability_list.append(probabilites)

    answer = []
    for key, value in class_probability.items():
        a = value
        for idx, val in enumerate(test_list):
            a *= probability_list[idx][key]
        answer.append(a)

    return answer.index(max(answer))
```

```python
answers = []
for item in X_test:
    answers.append(findAnswer(item))

answers
```

```python
matching = list(class_probability.keys())
matching
```
